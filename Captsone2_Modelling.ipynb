```python
X_train, X_test, y_train, y_test = train_test_split(data2, conv, test_size=.2, random_state=1)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
resultsDF = pd.DataFrame(columns=['n_estimators','MAE','MSE','RMSE','R^2','Average Absolute Error','Accuracy'])
for i in range(5,1200,100):
  regressor = RandomForestRegressor(n_estimators = i, random_state = 1,criterion='mse',n_jobs=-1)
  regressor.fit(X_train, y_train)
  y_test_pred =  regressor.predict(X_test)
  y_train_pred = regressor.predict(X_train)
  errors = abs(y_test_pred - y_test)
  mape = 100 * (errors / y_test)
  accuracy = 100 - np.mean(mape)
  regressor.feature_importances_
  print('n_estimators: %.3f' % (i))
  print('MAE train: %.3f,test: %.3f' % (mean_absolute_error(y_train,y_train_pred),mean_absolute_error(y_test,y_test_pred)))
  print('MSE train: %.3f,test: %.3f' % (mean_squared_error(y_train,y_train_pred),mean_squared_error(y_test,y_test_pred)))
  print('Root Mean Squared Error Train: %.3f,Root Mean Squared Error Test: %.3f' % (np.sqrt(mean_squared_error(y_train, y_train_pred)),np.sqrt(mean_squared_error(y_test, y_test_pred))))
  print('R^2 train: %.3f, test: %.3f' % (r2_score(y_train,y_train_pred),r2_score(y_test,y_test_pred)))
  print('Average absolute error:', round(np.mean(errors), 2), 'degrees.')
  print('Accuracy:', round(accuracy, 2), '%.')

  resultsDF = resultsDF.append({'n_estimators':i,'MAE':mean_absolute_error(y_test,y_test_pred),'MSE':mean_squared_error(y_test,y_test_pred),'RMSE':np.sqrt(mean_squared_error(y_test, y_test_pred)),'R^2':r2_score(y_test,y_test_pred),'Average Absolute Error':round(np.mean(errors), 2),'Accuracy':round(accuracy, 2)},ignore_index=True)

  plt.scatter(y_train_pred,y_train_pred - y_train,c='steelblue',edgecolors='white',marker='o',s=35,alpha=0.9,label='Training data')
  plt.scatter(y_test_pred,y_test_pred - y_test,c='limegreen',edgecolors='white',marker='o',s=35,alpha=0.9,label='Test data')
  plt.xlabel('Predicted values')
  plt.ylabel('Residuals')
  plt.legend(loc='upper left')
  plt.tight_layout()
  plt.show()

  features=data2.columns
  importances = regressor.feature_importances_
  indices = np.argsort(importances)

  plt.figure(1)
  plt.title('Feature Importances')
  plt.barh(range(len(indices)), importances[indices], color='b', align='center')
  plt.yticks(range(len(indices)), features[indices])
  plt.xlabel('Relative Importance')
  plt.show()

resultsDF
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

timeseriesDf=pd.read_csv('/content/gdrive/MyDrive/Springboard/Capstone2/Raw_Data/Cleaned_Dataset.csv')
timeseriesDf.reset_index(inplace=True)
timeseriesDf['Date'] = pd.to_datetime(timeseriesDf['Day'])
timeseriesDf['DOW'] = timeseriesDf['Date'].dt.day_name()
timeseriesDf['MonthName'] = timeseriesDf['Date'].dt.month_name()
timeseriesDf['Year'] = timeseriesDf['Date'].dt.year
timeseriesDf.set_index('Date', inplace=True)
timeseriesDf.head()
```

```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(window=24).mean()
    rolstd = timeseries.rolling(window=24).std()

    #Plot rolling statistics:
    fig = plt.figure(figsize=(12, 8))
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    
    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(timeseriesDf.Cost)
```

```python
timeseriesDf['cost_first-difference'] = timeseriesDf.Cost - timeseriesDf.Cost.shift(1)
test_stationarity(timeseriesDf['cost_first-difference'].dropna(inplace=False))
```

```python
timeseriesDf['cost_seasonal-difference'] = timeseriesDf.Cost - timeseriesDf.Cost.shift(12)
test_stationarity(timeseriesDf['cost_seasonal-difference'].dropna(inplace=False))
```

```python
timeseriesDf['cost_seasonal_first-difference'] = timeseriesDf['cost_first-difference'] - timeseriesDf['cost_first-difference'].shift(12)
test_stationarity(timeseriesDf['cost_seasonal_first-difference'].dropna(inplace=False))
```

```python
timeseriesDf.cost_log= timeseriesDf.Cost.apply(lambda x: np.log(x))  
test_stationarity(timeseriesDf.cost_log)
```

```python

timeseriesDf['log_cost_seasonal_difference'] = timeseriesDf.cost_log - timeseriesDf.cost_log.shift(12) 
test_stationarity(timeseriesDf['log_cost_seasonal_difference'].dropna(inplace=False))
```
